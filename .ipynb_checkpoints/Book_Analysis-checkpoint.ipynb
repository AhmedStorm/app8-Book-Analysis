# Find all sentences that contain "Love" word

```python
new_pat = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
find = re.findall(new_pat,book)
print(len(find))
```

```python
patt = re.compile("[a-zA-Z]+")
found = re.findall(patt,book.lower())
len(found)
```

```python
d = {}
for word in found:
    if word in d.keys():
        d[word] = d[word] +1
    else:
        d[word] = 1
        
tup = [(value, key) for key,value in d.items()]
sorted(tup, reverse = True)
        
```
